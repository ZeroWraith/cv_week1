# Matrix Addition

```python
from matrix_ops import Matrix
# --- Valid Addition ---
m1 = Matrix([[1, 2, 3],
             [4, 5, 6]])

m2 = Matrix([[7, 8, 9],
             [10, 11, 12]])

# Because we defined __add__, we can use the '+' operator!
m3 = m1 + m2
print(f"Successful addition result: {m3.data}")


# --- Invalid Addition ---
m4 = Matrix([[1, 2],
             [3, 4]])

print("\nadding matrices with different dimensions:")
try:
    result = m1 + m4
except ValueError as e:
    # Our custom error message gets printed!
    print(f"Caught an error: {e}")

```
